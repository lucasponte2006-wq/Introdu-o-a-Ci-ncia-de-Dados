>Identação Obrigatória

```
x = 5
if (x > 0):
  print("x é maior que zero")
else:
  print("x é menor ou igual a zero")
```

```
x = -5
if (x > 0):
  print("x é maior que zero")
```

```
x = -5
if (x > 0):
  print("x é maior que zero")
elif(x == 0):
  print("x é zero")
else:
  print("x é menor")
```

```
x = -6
if (x > 0):
  print("x é maior que zero")
elif(x < 0  and x > -4):
  print("x está entre 0 e -4")
elif(x <= -4 and x > -7):
  print("x está entre -4 e -7")
else:
  print("x é menor que -7")
```

###Funções

>Função que não retorna valor

```
def exibir_mensagem(parametro):
  print("Mensagem >> "+parametro)
```

```
exibir_mensagem('Olá!!!')
```

```
exibir_mensagem('Ola Mundo!')
```

```
def exibir_dados(nome, idade):
  print("Meu nome é {}".format(nome))
  print("Tenho {} anos de idade.".format(idade))
```

```
exibir_dados('Maria', 70)
```

```
exibir_dados('Paulo', 12)
```

>Função que retorna valor

```
def elevar_ao_quadrado(numero):
  return numero ** 2
```

```
elevar_ao_quadrado(2)
```

```
elevar_ao_quadrado(2)
elevar_ao_quadrado(25)
```

```
print(elevar_ao_quadrado(2))
print(elevar_ao_quadrado(25))
```

```
numero_quadrado = elevar_ao_quadrado(5)
```

```
numero_quadrado
```

```
for i in range(10):
  print(elevar_ao_quadrado(i))
```

```
x = 0
while elevar_ao_quadrado(x) <= 1000:
  print(x)
  x += 1
```
